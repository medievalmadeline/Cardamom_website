```python
#the list that we will make into a DF.
d = []
#for loop to get each page of cardamom recipes
for i in range(1,10):
    #grabbing the link for each page. There are 9 pages of cardamom recipes
    http_response = requests.get(f'https://cooking.nytimes.com/search?q=cardamom&page={i}')
    #changing to html
    html = http_response.text
    #building soup
    soup = BeautifulSoup(html)    
    #grabbing just the cardamom articles from the page
    articles = soup.select('.atoms_card__sPaoj')
    #for loop to grab each cardamom recipe and add the recipe and the link to the list(d)
    for article in articles:
        link = article.a.get('href')
        name = article.h3.text
        d.append(
            {
                'recipe': name,
                'link': f'https://cooking.nytimes.com{link}'
            
            }
        )



    
```

```python
#removing first item from list d 
d.pop(0)
```

```python
df = pd.DataFrame(d)
```

```python
recipe_d = []
for recipe in d:
    #getting each link from my list of recipes
    url = recipe['link']
    http_response = requests.get(url)
    html = http_response.text
    soup = BeautifulSoup(html)
    ingredients = soup.select('.ingredient_ingredient__lq70t')
    #making an empty list to put ingredients in
    ingred_list = []
    #counting ingredients to see if we have the right number 
    ingred_count = 0
    for ingred in ingredients: 
        ingred = ingred.text.lower().replace('teaspoon','')
        ingred_count +=1
        for food in food_list:
            if food.lower() in ingred:
                ingred_list.append(food)
    if len(ingred_list) == ingred_count:
        error = ''
    else:
        error = 'Error'
    print(error)
    recipe_d.append(
        {
            'recipe': recipe['recipe'],
            'link': url,
            'ingredients': ingred_list,
            'error': error
        }
    )

```
